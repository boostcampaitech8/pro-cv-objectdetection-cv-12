```python
import json
import random 
```

```python
with open("../../../dataset/train.json") as f:
    data=json.load(f)
images=data["images"]
annotations=data["annotations"]
categories=data["categories"]

random.shuffle(images)
val_ratio=0.2
val_count = int(len(images)*val_ratio)

val_images=images[:val_count]
train_images=images[val_count:]

val_img_ids=set([img["id"] for img in val_images])

val_ann=[ann for ann in annotations if ann["image_id"] in val_img_ids]
train_ann=[ann for ann in annotations if ann["image_id"] not in val_img_ids]

train_json={
    "images": train_images,
    "annotations": train_ann,
    "categories": categories
}
val_json={
    "images":val_images,
    "annotations": val_ann,
    "categories": categories
}

with open("new_train.json","w") as f:
    json.dump(train_json, f, indent=4)
with open("val.json","w") as f:
    json.dump(val_json, f, indent=4)
```

## split대로 이미지 옮기기

```python
import os
import json
import shutil
src_root = "/data/ephemeral/home/dataset/train"
dst_train = "/data/ephemeral/home/dataset_split15/train/train/"
dst_val = "/data/ephemeral/home/dataset_split15/valid/train/"
```

```python
with open("/data/ephemeral/home/dataset/new_train_split_15.json") as f:
    train_split=json.load(f)
    
    
with open("/data/ephemeral/home/dataset/new_val_split_15.json") as f:
    val_split=json.load(f)

    
train_files = [img["file_name"] for img in train_split["images"]]
val_files  = [img["file_name"] for img in val_split["images"]]
```

```python
print(len(val_files))
```

```python
def move_files(file_list, dst_dir):
    for filename in file_list:
        # image 이름 주의 "train/0001.jpg"
        base_name = os.path.basename(filename)

        src_path = os.path.join(src_root, base_name)

        if not os.path.exists(src_path):
            print(f"missing file: {src_path}")
            continue

        dst_path = os.path.join(dst_dir, base_name)

        os.makedirs(os.path.dirname(dst_path), exist_ok=True)

        shutil.copy(src_path, dst_path)  # 실제 이동하려면 move로 변경
        # shutil.move(src_path, dst_path)

    print(f"✔ copied {len(file_list)} files → {dst_dir}")


move_files(train_files, dst_train)
move_files(val_files, dst_val)
```
