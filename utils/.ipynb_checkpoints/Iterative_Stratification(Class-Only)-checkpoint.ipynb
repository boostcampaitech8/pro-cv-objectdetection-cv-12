```python
import json
import os
import numpy as np
from tqdm import tqdm
from collections import Counter, defaultdict

# ==========================
# 1. ÏÑ§Ï†ï
# ==========================
CLASSES = [
    "General trash", "Paper", "Paper pack", "Metal", "Glass",
    "Plastic", "Styrofoam", "Plastic bag", "Battery", "Clothing"
]

ROOT_DIR = "/data/ephemeral/home/dataset"
JSON_PATH = os.path.join(ROOT_DIR, "new_train.json")

# Í≤∞Í≥º Ï†ÄÏû• Í≤ΩÎ°ú
TRAIN_OUTPUT_PATH = os.path.join(ROOT_DIR, "new_train_split.json")
VAL_OUTPUT_PATH = os.path.join(ROOT_DIR, "new_val_split.json")

VAL_RATIO = 0.2

# ==========================
# 2. Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Class Ï†ïÎ≥¥ ÌååÏã±
# ==========================
print(f"Loading {JSON_PATH}...")
with open(JSON_PATH, 'r') as f:
    data = json.load(f)

images = data['images']
annotations = data['annotations']

# ID Îß§Ìïë
img_to_classes = defaultdict(list)

print("Pre-processing: Parsing Class info...")

for ann in annotations:
    img_id = ann['image_id']
    cat_id = ann['category_id']
    img_to_classes[img_id].append(cat_id)

# ==========================
# 3. Stratification ÏïåÍ≥†Î¶¨Ï¶ò (Class-Only)
# ==========================
def run_iterative_class_aware():
    print("Running Iterative Stratification (Class Only)...")
    
    # 1. Ï†ÑÏ≤¥ ÌÅ¥ÎûòÏä§ Îì±Ïû• ÎπàÎèÑ Í≥ÑÏÇ∞
    class_counts = Counter()
    for img_id in img_to_classes:
        class_counts.update(img_to_classes[img_id])
        
    # 2. Ï†ïÎ†¨: Ìù¨Í∑Ä ÌÅ¥ÎûòÏä§ Ìè¨Ìï® Ïù¥ÎØ∏ÏßÄ Ïö∞ÏÑ†
    sorted_ids = sorted([img['id'] for img in images],
        key=lambda x: min([class_counts[c] for c in img_to_classes[x]]) if img_to_classes[x] else float('inf'))

    val_ids = []
    
    train_counts_curr = defaultdict(int)
    val_counts_curr = defaultdict(int)
    
    for img_id in tqdm(sorted_ids):
        cats = img_to_classes[img_id]
        
        # Ïñ¥ÎÖ∏ÌÖåÏù¥ÏÖò ÏóÜÎäî Ïù¥ÎØ∏ÏßÄÎäî Í±¥ÎÑàÎúÄ (ÎÇòÏ§ëÏóê TrainÏúºÎ°ú Î∞∞Ï†ï)
        if not cats: continue
        
        score = 0
        for cat in cats:
            v = val_counts_curr[cat]
            t = train_counts_curr[cat]
            curr_total = v + t + 1
            curr_ratio = v / curr_total
            
            if curr_ratio < VAL_RATIO: score += 1
            else: score -= 1
        
        if score > 0:
            val_ids.append(img_id)
            for c in cats: val_counts_curr[c] += 1
        else:
            for c in cats: train_counts_curr[c] += 1
            
    return set(val_ids)

# Ïã§ÌñâÌïòÏó¨ Val ID Ï∂îÏ∂ú
val_ids_set = run_iterative_class_aware()

print(f"\nSplit Result -> Train: {len(images) - len(val_ids_set)} images, Val: {len(val_ids_set)} images")

# ==========================
# 4. JSON ÌååÏùº Î∂ÑÌï† Î∞è Ï†ÄÏû•
# ==========================
print("Splitting JSON data...")

# Í∏∞Î≥∏ Íµ¨Ï°∞ Î≥µÏÇ¨ (info, licenses, categoriesÎäî Í≥µÌÜµ)
train_data = {
    'info': data.get('info', {}),
    'licenses': data.get('licenses', []),
    'categories': data.get('categories', []),
    'images': [],
    'annotations': []
}
val_data = {
    'info': data.get('info', {}),
    'licenses': data.get('licenses', []),
    'categories': data.get('categories', []),
    'images': [],
    'annotations': []
}

# Ïù¥ÎØ∏ÏßÄ Î∂ÑÌï†
for img in tqdm(images, desc="Splitting Images"):
    if img['id'] in val_ids_set:
        val_data['images'].append(img)
    else:
        train_data['images'].append(img)

# Ïñ¥ÎÖ∏ÌÖåÏù¥ÏÖò Î∂ÑÌï†
for ann in tqdm(annotations, desc="Splitting Annotations"):
    if ann['image_id'] in val_ids_set:
        val_data['annotations'].append(ann)
    else:
        train_data['annotations'].append(ann)

# Ï†ÄÏû•
print(f"Saving to {TRAIN_OUTPUT_PATH}...")
with open(TRAIN_OUTPUT_PATH, 'w') as f:
    json.dump(train_data, f, indent=4)

print(f"Saving to {VAL_OUTPUT_PATH}...")
with open(VAL_OUTPUT_PATH, 'w') as f:
    json.dump(val_data, f, indent=4)

# ==========================
# 5. Í≤ÄÏ¶ù (ÎπÑÏú® ÌôïÏù∏)
# ==========================
print("\n[Final Check] Class Distribution in Generated JSONs:")

def get_counts(ann_list):
    counts = np.zeros(len(CLASSES))
    for ann in ann_list:
        counts[ann['category_id']] += 1
    return counts

t_counts = get_counts(train_data['annotations'])
v_counts = get_counts(val_data['annotations'])

print(f"{'Class':<15} | {'Train':<6} | {'Val':<6} | {'Val Ratio (%)':<15}")
print("-" * 55)
for i, cls_name in enumerate(CLASSES):
    t = t_counts[i]
    v = v_counts[i]
    total = t + v
    ratio = (v / total * 100) if total > 0 else 0
    print(f"{cls_name:<15} | {int(t):<6} | {int(v):<6} | {ratio:.2f}%")

print("\n‚úÖ All Done!")
```

```python
import json
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================
# 1. ÏÑ§Ï†ï
# ==========================
CLASSES = [
    "General trash", "Paper", "Paper pack", "Metal", "Glass",
    "Plastic", "Styrofoam", "Plastic bag", "Battery", "Clothing"
]

ROOT_DIR = "/data/ephemeral/home/dataset"
# Î∂ÑÏÑùÌï† ÌååÏùº Í≤ΩÎ°úÎì§
ORIGINAL_JSON = os.path.join(ROOT_DIR, "new_train.json")
TRAIN_SPLIT_JSON = os.path.join(ROOT_DIR, "new_train_split.json")
VAL_SPLIT_JSON = os.path.join(ROOT_DIR, "new_val_split.json")

SAVE_PATH = "class_distribution_comparison_v3.png"

# ==========================
# 2. Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è ÎπÑÏú® Í≥ÑÏÇ∞ Ìï®Ïàò
# ==========================
def get_class_ratios(json_path, dataset_name):
    if not os.path.exists(json_path):
        print(f"‚ö†Ô∏è {dataset_name} ÌååÏùºÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§: {json_path}")
        return None

    print(f"Loading {dataset_name}...")
    with open(json_path, 'r') as f:
        data = json.load(f)
    
    # Ïπ¥ÌÖåÍ≥†Î¶¨ IDÎ≥Ñ Ïπ¥Ïö¥Ìä∏
    counts = {i: 0 for i in range(len(CLASSES))}
    total_objects = 0
    for ann in data['annotations']:
        counts[ann['category_id']] += 1
        total_objects += 1
        
    # ÎπÑÏú®(%) Í≥ÑÏÇ∞ Î∞è Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
    stats = []
    if total_objects > 0:
        for i, cls_name in enumerate(CLASSES):
            ratio = (counts[i] / total_objects * 100)
            stats.append({
                "Class Name": cls_name,
                "Dataset": dataset_name,
                "Ratio (%)": ratio,
                "Count": counts[i] # (Ï∞∏Í≥†Ïö©)
            })
    return pd.DataFrame(stats)

# ==========================
# 3. Îç∞Ïù¥ÌÑ∞ ÌÜµÌï©
# ==========================
df_org = get_class_ratios(ORIGINAL_JSON, "Original")
df_train = get_class_ratios(TRAIN_SPLIT_JSON, "Train")
df_val = get_class_ratios(VAL_SPLIT_JSON, "Validation")

# ÏÑ∏ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ìï©ÏπòÍ∏∞ (Long Format)
df_final = pd.concat([df_org, df_train, df_val], ignore_index=True)

# ÌÅ¥ÎûòÏä§ ÏàúÏÑú Ïú†ÏßÄ
df_final["Class Name"] = pd.Categorical(df_final["Class Name"], categories=CLASSES, ordered=True)

# ==========================
# 4. ÏãúÍ∞ÅÌôî (Î≥¥Ïó¨Ï£ºÏã† ÏòàÏãú Ïä§ÌÉÄÏùº Ï†ÅÏö©)
# ==========================
print("\nüìä ÏãúÍ∞ÅÌôî ÏÉùÏÑ± Ï§ë...")

# Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞ Î∞è Ïä§ÌÉÄÏùº ÏÑ§Ï†ï
plt.figure(figsize=(16, 8))
sns.set_theme(style="whitegrid")

# ÏÉâÏÉÅ ÌåîÎ†àÌä∏ ÏÑ§Ï†ï (ÌöåÏÉâ, ÌïòÎäòÏÉâ, Ï£ºÌô©ÏÉâ Í≥ÑÏó¥)
custom_palette = {"Original": "#A9A9A9", "Train": "#87CEEB", "Validation": "#FFA500"}

# ÎßâÎåÄ Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞ (hueÎ°ú Îç∞Ïù¥ÌÑ∞ÏÖã Íµ¨Î∂Ñ)
ax = sns.barplot(
    data=df_final,
    x="Class Name",
    y="Ratio (%)",
    hue="Dataset",
    palette=custom_palette
)

# Ï†úÎ™© Î∞è ÎùºÎ≤® ÏÑ§Ï†ï
plt.title("Class Distribution Comparison (Original vs Train vs Validation)", fontsize=16, fontweight='bold')
plt.xlabel("", fontsize=12)
plt.ylabel("Ratio (%)", fontsize=12)
plt.xticks(rotation=45)
plt.ylim(0, df_final["Ratio (%)"].max() * 1.15) # YÏ∂ï ÏÉÅÎã® Ïó¨Ïú† Í≥µÍ∞Ñ ÌôïÎ≥¥

# ÎßâÎåÄ ÏúÑÏóê ÏàòÏπò ÌëúÏãú (ÏÜåÏàòÏ†ê Ï≤´Ïß∏ ÏûêÎ¶¨)
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{height:.1f}', 
                    (p.get_x() + p.get_width() / 2., height), 
                    ha='center', va='bottom', 
                    xytext=(0, 3), # ÌÖçÏä§Ìä∏ ÏúÑÏπò ÎØ∏ÏÑ∏ Ï°∞Ï†ï (ÏúÑÎ°ú 3Ìè¨Ïù∏Ìä∏)
                    textcoords='offset points',
                    fontsize=9, color='black')

# Î≤îÎ°Ä ÏúÑÏπò Ï°∞Ï†ï
plt.legend(title="Dataset", loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

# Ï†ÄÏû• Î∞è Ï∂úÎ†•
plt.savefig(SAVE_PATH, dpi=300, bbox_inches='tight')
print(f"‚úÖ Í∑∏ÎûòÌîÑ Ï†ÄÏû• ÏôÑÎ£å: {os.path.abspath(SAVE_PATH)}")
print("üëâ Í∑∏ÎûòÌîÑ Ìï¥ÏÑù: ÏÑ∏ Í∞ÄÏßÄ ÎßâÎåÄÏùò ÎÜíÏù¥Í∞Ä Í±∞Ïùò ÎπÑÏä∑Ìï†ÏàòÎ°ù, ÏõêÎ≥∏Ïùò Î∂ÑÌè¨Î•º Ïú†ÏßÄÌïòÎ©∞ Ïûò Î∂ÑÌï†Îêú Í≤ÉÏûÖÎãàÎã§.")
```
