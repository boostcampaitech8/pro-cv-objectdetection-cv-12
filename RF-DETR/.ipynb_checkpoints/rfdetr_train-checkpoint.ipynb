```python
from rfdetr import RFDETRMedium
import torch
```

```python
torch.cuda.empty_cache()
```

```python
data_dir = '../../../dataset_split15/' # data_dir 경로
```

```python
model=RFDETRMedium()

model.train(
    dataset_dir=data_dir,
    epochs=32,
    batch_size=4,
    grad_accum_steps=4,
    lr=1e-4,
    output_dir="./RF_DETR_ckpt",
    # early_stopping=True,
    # early_stopping_patience = 10,
    wandb=True,
    project='RFDETR',
    run='RFDETR_medium',
    checkpoint_interval = 5,
    do_random_resize_via_padding=True,
    resolution=768,
    multi_scale=True,
    use_ema=True,
    
)
```
